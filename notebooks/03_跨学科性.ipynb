```python
import mysql.connector

# 连接数据库
db = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="MySQLnewPassword2025",
    database="PAPER_sys"
)

# 创建游标
cursor = db.cursor()

# 查询用户表数据
cursor.execute("SELECT uname, email FROM USER")
for (uname, email) in cursor:
    print(f"用户名：{uname}，邮箱：{email}")

# 关闭连接
cursor.close()
db.close()
```
